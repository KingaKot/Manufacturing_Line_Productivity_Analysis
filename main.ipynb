## 5. Calculations

### Create full production timestamps and calculate batch duration

```python
# Create full production timestamps

productivity['Start_DateTime'] = pd.to_datetime(
    productivity['Date'].dt.strftime('%Y-%m-%d')
    + ' '
    + productivity['Start Time'].astype(str)
)

productivity['End_DateTime'] = pd.to_datetime(
    productivity['Date'].dt.strftime('%Y-%m-%d')
    + ' '
    + productivity['End Time'].astype(str)
)

# Handle batches crossing midnight
overnight_batches = productivity['End_DateTime'] < productivity['Start_DateTime']

productivity.loc[overnight_batches, 'End_DateTime'] = (
    productivity.loc[overnight_batches, 'End_DateTime']
    + pd.Timedelta(days=1)
)

# Calculate actual production time in minutes
productivity['Actual_Batch_Time_Min'] = (
    productivity['End_DateTime']
    - productivity['Start_DateTime']
).dt.total_seconds() / 60

print(productivity['Actual_Batch_Time_Min'].head())
```

## 6. Transform line_downtime Table

### Transform downtime data from wide to long format

```python
downtime_long = line_downtime.melt(
    id_vars='Batch',
    var_name='Factor',
    value_name='Downtime_Min'
)

downtime_long = downtime_long.dropna(
    subset=['Downtime_Min']
)

downtime_long.head(10)
```

```python
productivity.to_csv(
    'Productivity.csv',
    index=False
)

downtime_long.to_csv(
    'Downtime.csv',
    index=False
)

products.to_csv(
    'Product.csv',
    index=False
)

downtime_factors.to_csv(
    'Downtime_factor.csv',
    index=False
)
```
